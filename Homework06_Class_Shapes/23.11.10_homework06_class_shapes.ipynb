```python
#Parent Class: Shape
class Shape:
    def area():
        pass

#-----------------------------------------------------------------------------------
#Shape 0: Triangle:
class Triangle(Shape):
    def __init__(self, a, b, c): #a,b,c: 3 sides of a triangle
        self.a = a
        self.b = b
        self.c = c
        assert a+b>c 
        assert a+c>b
        assert b+c>a
        print(f"This triangle has the length of side 1 = {self.a}, side 2 = {self.b}, and side 3 = {self.c}")
    def area(self):
        p = (self.a + self.b + self.c)/2 #Heron formula to calculate the area of triangle when given 3 sides
        return (p * (p - self.a) * (p - self.b) * (p - self.c)) ** 0.5 #Heron formula to calculate the area of triangle when given 3 sides

triangle = Triangle(3,4,5)
print (f"The area of triangle is {triangle.area()}") #triangle.area relates to "triangle" in line triangle = class of Triangle(3,4,5)
print(triangle)
print ("\n")

#-----------------------------------------------------------------------------------
#Shape 1: Square:
class Square(Shape):
    def __init__(self, a):
        self.a = a
        print(f"This square has the length of each side = {self.a}")
    def area(self):
        return self.a * self.a  

square = Square(5)
print(f"The area of this square is {square.area()}")

print ("\n")

#-----------------------------------------------------------------------------------
#Shape 2: Rectangle:
class Rectangle(Shape):
    def __init__(self, a, b):
        self.a = a
        self.b = b
        print(f"This rectangle has the length of side 1 = {self.a}, and side 2 = {self.b}")
    def area(self):
        return self.a * self.b
    
rectangle = Rectangle(3,4)
print(f"The area of this rectangle is {rectangle.area()}")

print ("\n")

#-----------------------------------------------------------------------------------
#Shape 3: Circle
import math
class Circle(Shape):
    def __init__(self, r):
        self.r = r
        print(f"This circle has the radius = {self.r}")
    def area(self):
        return math.pi * self.r**2 #import math library so that I can have the exactly correct pi number)

circle = Circle(4)
print(f"The area of this circle is {round(circle.area(),2)}") #round function: round (value, number you want to round) 
```
